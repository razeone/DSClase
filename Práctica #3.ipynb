```python
import pandas as pd
df = pd.read_csv("datasets/mex_clean.csv")
```

```python
import seaborn as sns
ax = df.plot.scatter(y = 'pickup_latitude', x = 'pickup_longitude')
ax.set_xlim(-99.191076,-99.088220)
ax.set_ylim(19.355317, 19.437556)
```

```python
df['pickup_hour'] = pd.to_datetime(df.pickup_datetime).apply(lambda x:x.hour)
```

```python
df.head()
```

```python
import pylab as plt
g = sns.PairGrid(df, y_vars = ['pickup_latitude'], x_vars = ['pickup_longitude'], hue = 'pickup_hour' )
ax = g.map(plt.scatter, alpha = 0.5)
axes = ax.axes
axes[0,0].set_xlim(-99.191076, -99.0882209)
axes[0,0].set_ylim(19.355317, 19.437556)
plt.legend()
```
