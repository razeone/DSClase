```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('datasets/mpg.csv')
```

```python
df.head(5)
```

```python
df.plot.scatter(x = 'horsepower', y = 'mpg', c = 'weight')
```

```python
import pylab as plt
df.plot.scatter(x = 'horsepower', y = 'mpg')
plt.xlabel('Horsepower')
plt.ylabel('Miles per Galon')
```

```python
tips = pd.read_csv('datasets/tips.csv')
```

```python
tips.head(2)
```

```python
tips.hist()
```

```python
tips['tip'].hist(bins = 4)
plt.xlabel('Tips')
plt.ylabel('Count')
plt.title('My awesome chart')
```

```python
g = sns.FacetGrid(tips, row = 'sex', col = 'time')
g.map(plt.hist, 'total_bill')
```

```python
g = sns.FacetGrid(df, col = 'origin')
g.map(plt.scatter, 'mpg', 'horsepower')
plt.savefig('myScatterplot.png')
```
